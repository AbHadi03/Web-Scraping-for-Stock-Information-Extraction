```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
# Function to extract Product Title
def get_title(new_Soup):

    try:
        # Outer Tag Object
        title = new_Soup.find("span", attrs={'class':'stock-info-heading stock-info-ho fw700'})

        # Inner NavigatableString Object
        title_value = title.text

        # Title as a string value
        title_string = title_value.strip()

    except AttributeError:
        title_string = ""

    return title_string

# Function to extract Product Title
def get_nse_name(new_Soup):

    try:
        # Outer Tag Object
        NSE_name = new_Soup.find("span", attrs={'class':'m-l-1 fs075rem gr stock-info-details'})

        # Inner NavigatableString Object
        nse_value = NSE_name.text

        # Title as a string value
        nse_string = nse_value.strip()

    except AttributeError:
        nse_string = ""

    return nse_string
```

```python
if __name__ == '__main__':

    # add your user agent
    HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/125.0.0.0 Safari/537.36', 'Accept-Language': 'en-US, en;q=0.5'})

    # The webpage URL
    URL = "https://trendlyne.com/equity/sector/5/hardware-technology-equipment/"

    # HTTP Request
    webpage = requests.get(URL, headers=HEADERS)

    # Soup Object containing all data
    soup = BeautifulSoup(webpage.content, "html.parser")

    # Fetch links as List of Tag Objects
    links = soup.find_all("a", attrs={'class':'stockrow'})

    # Store the links
    links_list = []

    # Loop for extracting links from Tag Objects
    for link in links:
            links_list.append(link.get('href'))

    d = {"Title":[], "NSE NAME":[]}

    # Loop for extracting product details from each link
    for link in links_list:
        new_webpage = requests.get("https://trendlyne.com" + link, headers=HEADERS)

        new_Soup = BeautifulSoup(new_webpage.content, "html.parser")

        # Function calls to display all necessary product information
        d['Title'].append(get_title(new_Soup))
        d['NSE NAME'].append(get_nse_name(new_Soup))

    stock_df = pd.DataFrame.from_dict(d)
    stock_df['Title'].replace('', np.nan, inplace=True)
    stock_df = stock_df.dropna(subset=['Title'])
    stock_df.to_csv("Hardware_Technology.csv", header=True, index=False)
```
